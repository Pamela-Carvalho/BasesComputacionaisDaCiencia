## Planilhas com outros delimitadores de dados

Podemos utilizar outros delimitadores em vez de `,`. Um outro delimitador muito utilizado é o `;`.

O ``pandas`` considera vírgula como delimitador padrão de dados. Se você for ler uma planilha que usa outro delimitador de dados que não seja vírgula, você tem que especificar o delimitador quando usar ``read_csv``.

Veja a seguir como fazer isso. Usamos `sep=' ; '`.

Note que e a mesma planilha com que estamos trabalhando, mas usando o delimitador de dados ponto-e-vírgula ``;``.

````python
filmes2_csv = "4.1 Filmes2.csv"

f = open(filmes2_csv, "w", encoding='utf-8')
f.write("""Título;Ano;Duração;Nota
Star Wars;1977;121;87
Harry Potter e a Pedra Filosofal;2001;152;75
O Senhor dos Anéis: a Companhia do Anel;2001;178;88
Os Vingadores;2012;143;84
Star Trek;2009;127;80
X-men;2000;104;74
Os Guardiões da Galáxia;2014;121;82
A Origem;2010;148;88
The Matrix;1999;136;87
O Exterminador do Futuro;1984;107;81
""")
f.close()

df = pd.read_csv(filmes2_csv, sep=';')
df
````

Veja o que acontece nesta planilha se você não especificar o ``;``:

````python
df = pd.read_csv(filmes2_csv)
df
````

O arquivo é lido, mas o ``pandas`` considera que ele tem apenas uma coluna.

Caso você queira utilizar o caractere *delimitador* dentro de uma célula, você pode usar `quote`.

Veja o exemplo completo abaixo:

````python
exemplo_csv = "4.1 ExemploCompleto.csv"

f = open(exemplo_csv, "w", encoding='utf-8')
f.write("""Id:Nome:Descrição
1:Fulano:Alguma descrição # Aqui é um comentário e será ignorado
2:Beltrano:*Texto dentro da **quote** delimitado por ***
3:Cicrano:*O caractere # não funciona dentro da quota* # Mas aqui fora funciona!
""")
f.close()

df = pd.read_csv(exemplo_csv, sep=':',quotechar='*',comment='#')
df
````

**Observações:**
- Utilizamos `:` como separador neste exemplo
- O caractere padrão para ***quote*** é o `"`, mas alteramos neste exemplo para `*`
    - Para utilizar o caractere definido como ***quote*** dentro da própria ***quote***, basta duplicá-lo. No exemplo: `**` representa um único `*`.
- Embora não usual, podemos incluir comentários. Definimos o caractere utilizado como comentário em `comment`, e tudo que está a direita deste caractere será ignorado
    - O caractere definido em `comment` não tem efeito se estiver dentro de uma ***quote***

# Ordenação: comando `sort_values(by, ascending)`

OBS:

by= nome do atributo, ou lista de nomes de atributos

Exemplo: by=["Nota","Ano"]

ascending = variavel Booleana ou lista de variáveis Booleanas

Exemplo: ascending=[False, True]

Uma das operações frequentemente utilizadas em planilhas é a operação de **ordenação**.

Veja **como ordenar em ordem crescente de "Ano" (do mais antigo para o mais recente).**

OBS: **Por padrão, parâmetro ``ascending`` é considerado como ```True```.**

````python
df = pd.read_csv(filmes_csv)
df.sort_values(by="Ano")
# ordenar em ordem crescente de "Ano" (do mais antigo para o mais recente).
````

**E se quisermos ordenar em ordem decrescente de "Ano" (do mais recente para o mais antigo)?**

**Adicionamos mais um parâmetro** quando usamos o sort_values: o nome do parâmetro é **```ascending``` e vamos colocar o valor ```False``` nele.**

**Por padrão, esse parâmetro ``ascending`` é considerado como ```True```.**

````python
df.sort_values(by="Ano", ascending=False)
# ordenar em ordem decrescente de "Ano"
````

## Ordenação com critério de desempate

**Note que "A Origem"	e "O Senhor dos Anéis" têm a mesma Nota 88, mas foram lançados em anos diferentes.**

Suponha que queiramos **ordenar por Nota (ordem crescente) e** utilizar como **critério de desempate o Ano (ordem crescente)**.

````python
df.sort_values(by=["Nota","Ano"])
# ordena por Nota (ordem crescente) e utiliza como critério de desempate o Ano (ordem crescente)
````

Note o uso de ``[`` e ``]`` acima, com "Nota" e "Ano" separados pela vírgula ``,``.

**E se quisermos que seja em ordem decrescente de Nota, mas ordem crescente de Ano? Utilizamos o ``ascending`` novamente, mas agora com dois valores:** False e True



````python
df.sort_values(by=["Nota","Ano"], ascending=[False, True])
# ordena por Nota (ordem decrescente) e utiliza como critério de desempate o Ano (ordem crescente)
````

# Modificar a planilha ou criar uma cópia?
Quando usamos o `sort_values`, a planilha em `df` não é modificada.

Confira!

````python
df.sort_values(by="Ano")
````

````python
df
````

Quando executamos
```
df.sort_values(by="Ano")
```
* Criamos uma cópia de df e trabalhamos na cópia
* Podemos guardar essa cópia em outra variável, por exemplo, df_ordenada:

````python
df_ordenada = df.sort_values(by = "Ano")
df_ordenada
````

Note que os índices das linhas são preservados.
Isto é útil para manter a coerência da tabela em diferentes etapas de processamento.
Como exemplo, ao aplicar diversas operações na tabela, a identificação da linha original é preservada.

Caso quisermos refazer o índice, precisamos chamar `reset_index`.

**Atenção:** Ao executar esta função, a relação da tabela atual e a original se perde.

Veja o exemplo abaixo:

````python
df_ordenada.reset_index()
````

Mas e se quiséssemos modificar a própria planilha ao invés de criar uma nova?

Podemos colocar a cópia na própria df:

````python
df = df.sort_values(by="Ano")
df
````

Uma outra maneira é a seguinte, usando inplace=True :

````python
df.sort_values(by="Ano", inplace=True)
df
````

### Qual a diferença entre essas duas opções?

### Por que temos duas maneiras de fazer a mesma atividade/tarefa?

Essas duas opções atingem o mesmo efeito: a planilha em df agora se encontra ordenada.

Mas os passos realizados são diferentes:

```df = df.sort_values(by="Ano")```
* Cria uma cópia de df
* Ordena a cópia de df
* Transfere a cópia para df

```df.sort_values(by="Ano", inplace=True)```
* Ordena a própria df sem retornar uma cópia

A primeira opção precisa de mais memória do computador por causa da cópia. Isso não é problema para planilhas pequenas, mas poderia ser se a planilha fosse muito grande.

# Filtros: comando `query("condição")`

Filtros são ferramentas muito importantes para a visualização de dados, pois nos **permitem ver apenas as entradas que satisfazem alguma condição**.

Veja df novamente:

````python
df
````

Suponha que você queira implementar um filtro para verificar os **filmes com duração menor do que 125 minutos**. Se você executar o seguinte código, qual resultado  será apresentado?

````python
df.query("Duração < 125")
````

E se quisermos olhar **apenas filmes do ano 2000**?


````python
df.query("Ano == 2000")
````

E **apenas o filme Star Wars**? Observação: Para fazer uma condição de igualdade, use ``==`` e não ``=``

````python
df.query("Título == 'Star Wars'")
````

Note que no comando acima 'Star Wars' aparece entre  **aspas simples**  (para indicar que é um texto).

E se quisermos olhar os **filmes com menos de 125 minutos a partir do ano 2000**?

Para isso combinamos as duas condições usando um ``and``:

````python
df.query("Duração <125 and Ano >= 2000")
````



O ``and`` representa um "e lógico", quando queremos que as duas condições valham.

Ou seja, a condição 1 e a condição 2 valham ao mesmo tempo.

E se quiséssemos ver os filmes que satisfazem **pelo menos uma das condições**?

Por exemplo, você quer ver um **filme com menos de 125 minutos, mas, se for mais longo, você topa desde que ele seja a partir dos anos 2000**.

Nesse caso, usamos um ``or``, que representa um "ou lógico"

````python
df.query("Duração < 125 or Ano >= 2000")
````

O "ou-lógico" é um pouco diferente do "ou" que usamos em Português usualmente.

O "ou-lógico" é o que chamamos de "ou-inclusivo" que significa que ou a primeira condição vale, ou a segunda condição vale, ou as duas valem.

## Agrupar - comando `groupby`

A ferramenta de agrupamento permite agregar dados e analisá-los, de acordo com algum dos campos.

Por exemplo, Star Wars e The Matrix são os filmes no grupo de nota 87 e têm duração 121 e 136, respectivamente. Então, a média é (121+136)/2 = 257/2 = 128.5.

* Contando os títulos em cada grupo

# Planilhas csv que tenham a separação decimal de números reais marcada por vírgula [,] e não o ponto.



Em muitos casos precisamos trabalhar com planilhas que utilizam vírgulas como separação decimal.

A primeira solução seria substituir todas as `,` por `.`, mas esta solução não é interessante quando acessamos uma base de dados externa.

Como exemplo, se queremos obeter a cotação do Dollar diário atualizado, podemos acessar a tabela diretamente no site do Banco Central do Brasil.

Entretanto, esta tabela usa o delimitador de dados ponto-e-vírgula ``;`` e o separador de casas decimais vírgula ``,``.

A célula abaixo lê as cotações em formato de separador decimal com `,` e separador com `;`:

````python
df = pd.read_csv("https://api.bcb.gov.br/dados/serie/bcdata.sgs.10813/dados?formato=csv", sep=';', decimal=",")
````


Neste caso, ao invés de somente

df = pd.read_csv("link", sep = ';')

deve ser adicionar **decimal = ","** como o último parâmetro,

**para indicar** **para o Python que o separador DECIMAL na tabela é com VÍRGULA e não ponto** (o Python espera ponto). Assim, o comando fica:

df = pd.read_csv("link", sep=";", **decimal = ","**)

Alguns exemplos:

````python
print(df.head(9))

print(df.tail(9))
````

Note que a data é representada por um texto nesta planilha, e o `pandas` não consegue interpretar.

No exemplo abaixo, estamos filtrando apenas as cotações de 2023.

````python
df.query("data >= '2023-01-01' and data <= '2023-12-31'")
````

**Obs.:** O formato "YYYY-MM-DD" (Ano-Mês-Dia) segue o padrão ISO 8601, e é muito utilizado como representação universal.

Note que não há resultados para a `query` acima.

Podemos fazer o `pandas` interpretar a data corretamente rodando a célula abaixo.

````python
# Este comando informa o formato utilizando para data, atualizando a coluna 'data'
df['data'] = pd.to_datetime(df['data'], format="%d/%m/%Y") # https://docs.python.org/3/library/datetime.html#strftime-and-strptime-behavior
````

Após este comando, a coluna `data` de `df` é atualizado, interpretando corretamente a data.

Agora, as `query` com datas funcionam:

````python
# Cotações 2023
df.query("data >= '2023-01-01' and data <= '2023-12-31'")
````

Uma vez reconhecido a coluna data corretamente, podemos aplicar os filtros é fórmulas já aprendidas.

````python
# Cotação média de 2023
df.query("data >= '2023-01-01' and data <= '2023-12-31'").mean(numeric_only=True)
````

Agora vamos obter o maior índice do Dollar após o ano de 2000:

````python
# Cotação máxima após 2000
df.query("data >= '2000-01-01'").max()
````

Embora o valor máximo da cotação esteja correto, a data é igual à cotação mais recente, e não maior.
Esta informação está correta, pois o `pandas` retornou o valor de cada coluna.

Entretanto, para obter a data da maior cotação, temos que alterar a forma acima e obter a linha da maior cotação, e então obter a data.

Cada linha é definida por um `index`. Então podemos utilizar a função `idxmax` para obter o índice da cotação máxima.

O código abaixo retorna o valor da cotação máxima e a respectiva data corretamente.

````python
# Cotação máxima após 2000
idx = df.query("data >= '2000-01-01'")[['valor']].idxmax()
df.iloc[idx]
````

Podemos formatar um texto com o resultado, conforme a célula abaixo.

````python
# Cotação máxima após 2000
idx = df.query("data >= '2000-01-01'")[['valor']].idxmax()[0]
print("A maior cotação foi de", df.at[idx,'valor'], "em", df.at[idx,'data'].strftime("%d/%m/%Y"))
````

**Obs.:** Para imprimir a data, foi utilizada a função `strftime` para formatar a data como dia/mês/ano.
