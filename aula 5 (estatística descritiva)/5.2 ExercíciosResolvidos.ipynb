Em especial, vamos lembrar o exemplo:


```python
C = 120.994901
print("O valor da variável C é %.2f"% C)
```

**Quando precisamos imprimir mais de uma mensagem e imprimir o valor de mais de uma variável formatada, podemos usar o seguinte:**


```python
C = 120.994901
F = 20.6547
print("O valor da variavel C = %.2f, e o valor da variavel F = %.2f"%(C,F))
```

Ou seja, para imprimir duas mensagens e duas variáveis formatadas (com, por exemplo, duas casas decimais), usamos o comando:

**print("mensagem_1 %.2f  mensagem_2 %.2f" %(variavel_1,variavel_2))**

Observe que as variáveis a serem impressas estão entre aspas e separadas pela(s) vírgula(s).

O mesmo comando pode ser usado para imprimir mais de duas mensagens e variáveis correspondentes.

### Exercício 1

Considere a planilha em:

https://drive.google.com/u/1/uc?id=182_Z-bL7n9itkupECYhIOEGuIjfQbF0Z&export=download

Calcule a média e a mediana da temperatura.

Interprete o seu resultado.

```python
import pandas as pd
df = pd.read_csv("https://drive.google.com/u/1/uc?id=182_Z-bL7n9itkupECYhIOEGuIjfQbF0Z&export=download")
valor_media = df["temperatura"].mean()     # calculando media
valor_mediana = df["temperatura"].median() # calculando mediana
print("Os valores sao:")                                              # imprimindo a mensagem
print("Media = %.2f, Mediana = %.2f"%(valor_media,valor_mediana))     # imprimindo os valores calculados com duas casas decimais
```

Note que o valor de média e mediana são próximos neste caso. Então, a média descreve bem o que está acontecendo.

De fato, na planilha ordenada, podemos ver que 14 temperaturas estão acima de média e 15 temperaturas estão abaixo da média:


```python
df.sort_values(by="temperatura")
```

### Exercício 2

Considere a planilha em:

https://drive.google.com/u/1/uc?id=182_Z-bL7n9itkupECYhIOEGuIjfQbF0Z&export=download

Calcule os percentis: 25ᵒ, 55ᵒ, 75ᵒ.

```python
import pandas as pd
import numpy as np    # precisamos de biblioteca numpy para calcular Percentil
df = pd.read_csv("https://drive.google.com/u/1/uc?id=182_Z-bL7n9itkupECYhIOEGuIjfQbF0Z&export=download")
p25 = np.percentile(df["temperatura"],q=25) # calculamos os Percentis e atribuimos valores as variaveis
p55 = np.percentile(df["temperatura"],q=55)
p75 = np.percentile(df["temperatura"],q=75)
print("Os Percentis calculados: 25ᵒ = %.1f, 55ᵒ= %.1f, 75ᵒ = %.1f"%(p25,p55,p75)) # imprimimos os Percentis calculados com 1 casa decimal
```

### Exercício 5
Considere a planilha em:

https://drive.google.com/u/1/uc?id=182_Z-bL7n9itkupECYhIOEGuIjfQbF0Z&export=download

Escreva instruções para ler o nome de uma coluna da tabela (período ou temperatura), obter e imprimir a(s) moda(s) no conjunto de dados da coluna lida.

```python
import pandas as pd
df = pd.read_csv("https://drive.google.com/u/1/uc?id=182_Z-bL7n9itkupECYhIOEGuIjfQbF0Z&export=download")
coluna = input("Digite o nome da coluna (periodo ou temperatura:) ")          # recebendo o nome do atributo/coluna
aux = df[coluna].mode()                                                       # calculando  a(s) moda(s) da coluna lida e atrubuindo o(s) valor(es) na variavel aux
print("A(s) moda(s) da coluna",coluna,":")                                    # imprimimos a mensagem e o nome da coluna lida
print(aux)                                                                    # imprimimos a(s) moda(s)
```

Note que no caso quando a coluna lida é "periodo", são impressas 30 modas, pois todos os números de períodos (de cada linha) são diferentes. Ainda, os valores de períodos são impressos em ordem crescente.

OBS:

Pela definição, se todos os elementos de um conjunto são diferentes, a moda NÃO EXISTE.

Mesmo assim, Python neste caso vai imprimir todos os elementos na ordem crescente, como no exemplo quando a coluna lida é "periodo".

O arquivo https://www.dropbox.com/s/6qwgc2dgyat96qd/fertilidade.csv?dl=1 contém informações de fertilidade e uso de contraceptivos em vários países, grupado por região. Esse arquivo usa ponto e vírgula como delimitador/separador dos dados, e ponto como indicador decimal. Calcule a variância da fertilidade. Considere apenas linhas para quais valor Contraceptivos é menor que 30.


Lembre-se (link abaixo) que ``pandas`` considera vírgula como delimitador padrão de dados. Se você for ler uma planilha que usa outro delimitador de dados que não seja vírgula, você tem que especificar o delimitador quando usar ``read_csv``. Nesse caso, usamos sep=' ; '

https://colab.research.google.com/drive/1-LbDUJYrBx6BNheUu36olNDe0vADjCqy?usp=sharing#scrollTo=W2TD_aLRgAP9

```python
import pandas as pd
df = pd.read_csv("https://www.dropbox.com/s/6qwgc2dgyat96qd/fertilidade.csv?dl=1", sep=";") #  sep=";" eh segundo parametro pois arquivo usa ponto e vírgula como delimitador dos dados

df_filtrada = df.query("Contraceptivos < 30")                                       # filtragem
variancia = df_filtrada["Fertilidade"].var()                                        # calculo de variancia de fertilidade na tabela ja filtrada
print("Varianacia considerando valor Contraceptivos < 30 eh")                       # impressao de resultado: mensagem
print("%.2f" %variancia)                                                            # impressao de resultado: valor com 2 casas decimais
```

Esse arquivo usa vírgula `,`como separador/delimitador dos dados e vírgula `,` como indicador decimal (ou seja, separação decimal de números reais é marcada por vírgula e não o ponto, como pode se ver na tabela acima).

Calcule a média de atributo "Social support" para um país lido do teclado com o comando input(). Considere esse cálculo apenas para Anos anteriores à 2017.


```python
import pandas as pd
# O arquivo tem vírgula como indicador/delimitador decimal, e não ponto (o Python espera ponto).
# Portanto,  ADICIONAMOS decimal = "," como o último parâmetro na hora de carregar (ler) a planilha:
df = pd.read_csv("https://www.dropbox.com/s/c598s62iq668cz4/Chapter2OnlineData-9.csv?dl=1",decimal=",")

Pais_Lido=input()                                                   # leitura de nome de pais (por exemplo, Afghanistan, Albania, Brazil...)
df_filtrada = df.query("Country == '"+Pais_Lido+"' and Year < 2017")# criamos uma cópia de planilha df (chamada df_filtrada) que contém:
                                                                    # linhas cujo  atributo "Country" é igual ao Pais_Lido  e também ano < 2017
print("Media de Suporte Social no pais",Pais_Lido,"e ano < 2017:")  #imprimimos mensagem
print(df_filtrada["Social support"].mean())                         # calucalmos e imprimimos a media de atributo "Social support" da tabela df_filtrada
```

**IMPORTANTE:**

Observe a sintaxe na sexta linha no programa acima

**Country == '  " +Pais_Lido+ "   '**

Quando desejamos filtrar dados usando um atributo, por exemplo `Country`, cujos elementos não são números, mas sim, um **texto** (no caso, nomes de países), fazemos:

1) Entramos com o nome de país (elemento de atributo) via teclado com comando `input()` - por exemplo, Afghanistan, Albania...

2) Atribuímos este valor lido  à uma variável, chamada, como no exemplo acima, `Pais_Lido`

3) Usamos:

**Country == ' " +Pais_Lido+ " '** para filtrar os elementos da coluna `Country` iguais ao nome/elemento digitado via teclado `input() `

Ou seja, no caso geral:

 **nome_atributo == ' " +nome_variável+" '**
