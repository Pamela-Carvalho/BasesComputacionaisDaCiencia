### Exercício

Faça um gráfico para a função $f(x) = x^3-3x^2+100$ para $x$ entre $-5$ e $2$.

```python
x = np.arange(-5, 2.1, .1)
y = x ** 3 - 3 * x ** 2 + 100
plt.plot(x, y)
plt.grid(True)
```

### Exercício

O algoritmo para fazer o gráfico da função $f(x) = \text{sen}(x)\cdot(\text{sen}(x)+\text{cos}(x))^2$ para $x$ entre $-5$ e $5$ está correto?

```python
x = np.arange(-5,5.1, 0.1)
y = np.sin(x)*(np.sin(x)+np.cos(x))**2
plt.plot(x,y)
# sim, está correto
```

### Exercício

Faça dois gráficos juntos:

- $f_1(x) = \cos(x)$ para $x$ de $-25$ a $25$.
- $f_2(x) = \sqrt{226-x^2}$  para $x$ de $-15$ a $15$.

```python
x1 = np.arange(-25, 25.1, 0.1)
y1 = np.cos(x1)

x2 = np.arange(-15, 15.1, 0.1)
y2 = np.sqrt(226 - x2 ** 2)

plt.plot(x1, y1)
plt.plot(x2, y2)
```

### Exercício
Modifique o título, eixos, os rótulos, cores e estilos das linhas.

```python
x = np.arange(-10,10,1)
y1 = x*x
y2 = -x*x
y3 = x
plt.plot(x,y1,'cyan',label='ciano')
plt.plot(x,y2,'purple',label='roxo')
plt.plot(x,y3,'yellow',label='amarelo')
plt.title("Meu gráfico lindo")
plt.grid(True)
plt.legend("legenda")
plt.xlabel("eixo x")
plt.ylabel("eixo y")
```
